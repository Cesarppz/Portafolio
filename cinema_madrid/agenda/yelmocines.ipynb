```python
data_horas = []
n = 0
for i in range(20000):
    print(fechas[i])
    fechas_button.click()
    respuesta = input('')
    if respuesta == 'E':
        break
    for name in cines_names:
        
        titles = driver.find_elements_by_xpath(f'//div[@data-cinema="{name}"]//h3/a')
            
        titles = [x.text for x in titles]
        images = driver.find_elements_by_xpath(f'//div[@data-cinema="{name}"]//figure[@class="col4"]/a/img')
        images = [x.get_attribute('src') for x in images]
        new_images = []
        try:
            for idc, image in enumerate(images):
                image_name = download_images.download(image,nombre_del_lugar='cines_yelmo',idx=idc+1,len_links=n)
                new_images.append(image_name)
            n += 1
        except:
            image_name = None
        
        horas = driver.find_elements_by_xpath(f'//div[@data-cinema="{name}"]//div[@class="cf"]')
        time = [x.text.split('\n')[1] for x in horas]
        
        categoria = [x.text.split('\n')[0].split(' ')[-1] for x in horas]
        categoria = [x.lower().replace('español','cine').capitalize() for x in categoria ]
        
        names = [name.capitalize() for i in titles]
        single_day = [fechas[i] for x in titles]
        
        data_horas.append({'title': titles, 'horas':time, 'nombre_cine':names, 'categoria':categoria, 'Fecha':single_day, 'Image':new_images})
```
